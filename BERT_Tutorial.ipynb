```
import pandas as pd
import numpy as np

# load training data
train = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/train_E6oV3lV.csv', encoding='iso-8859-1')
train.shape
train=train.head(10000)
```

```
import re

# clean text from noise
def clean_text(text):
    # filter to allow only alphabets
    text = re.sub(r'[^a-zA-Z\']', ' ', text)
    
    # remove Unicode characters
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # convert to lowercase to maintain consistency
    text = text.lower()
       
    return text

train['clean_text'] = train.tweet.apply(clean_text)
```

```
	from sklearn.model_selection import train_test_split

# split into training and validation sets
X_tr, X_val, y_tr, y_val = train_test_split(train.clean_text, train.label, test_size=0.25, random_state=42)
# tr is training, val is validation

print('X_tr :',X_tr)
print('X_tr shape:',X_tr.shape)
print('y_tr:',y_tr)
# print('y_tr shape:',y_tr.shape)
print("")
print('X_val :',X_val)
print('X_val shape :',X_val.shape)
print('y_val:',y_val)
# print('y_val shape:',y_val.shape)
```

```
# get the embedding for train and val sets
print("trainign beginning")
X_tr_bert = get_embeddings(X_tr.tolist())
print (len(X_tr_bert))
print("X_tr done")
X_val_bert = get_embeddings(X_val.tolist())
print (len(X_val_bert))
```

```
from sklearn.linear_model import LogisticRegression

# LR model
model_bert = LogisticRegression(max_iter=100000)
# training i.e. learning relationship between tweets and their hate speech flags. But instead of the tweet string itself, we are using BERT's embedding
model_bert = model_bert.fit(X_tr_bert, y_tr)
print(model_bert)
# predict
pred_bert = model_bert.predict(X_val_bert)
```

```
from sklearn.metrics import accuracy_score

print(accuracy_score(y_val, pred_bert))
```
